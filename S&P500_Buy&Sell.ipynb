```python
sp500 =yf.Ticker("^GSPC")
```

```python
sp500 =sp500.history(period="max")
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500["Tomorrow"]=sp500["Close"].shift(-1)
sp500
```

```python
sp500["Target"]=sp500["Close"]<sp500["Tomorrow"]
sp500
```

```python
sp500["Target"]=(sp500["Close"]<sp500["Tomorrow"]).astype(int)
sp500
```

```python
sp500=sp500.loc["2000-04-10":].copy()
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier 
model = RandomForestClassifier(n_estimators=400, min_samples_split=100,random_state=1)
```

```python
test =sp500.iloc[-1000:]
train =sp500.iloc[:-1000]
predictors=["Open","Close","High","Low","Volume"]

model.fit(train[predictors],train["Target"])
```

```python
from sklearn.metrics import precision_score
import pandas as pd

predicted=model.predict(test[predictors])
```

```python
predicted=pd.Series(predicted, index=test.index)
predicted
```

```python
combined=pd.concat([test["Target"],predicted],axis=1)
combined
```

```python
precision_score(test["Target"],predicted)

```

```python
#fuction for backtesting

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    predicted = model.predict(test[predictors])
    predicted = pd.Series(predicted, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], predicted], axis=1)
    return combined



def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    
    return pd.concat(all_predictions)
  




  
```

```python
predictions=backtest(sp500,model,predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
thresholds =[2,5,60,250,1000]
new_predictors=[]
for threshold in thresholds:
  rolling_avgs= sp500.rolling(threshold).mean()  #we are trying to determine if the price today is greater than the previous threshold price
  ratio_column=f"Close_Ratio{threshold}"
  sp500[ratio_column]=sp500["Close"]/rolling_avgs["Close"]
  trend_column = f"Trend{threshold}"
  sp500[trend_column]=sp500.shift(1).rolling(threshold).sum()["Target"]

  new_predictors +=[ratio_column,trend_column]
```

```python
model = RandomForestClassifier(n_estimators=2000, min_samples_split=30,random_state=1)

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    predicted = model.predict_proba(test[predictors])[:,1]
    predicted[predicted >=.6] = 1
    predicted[predicted <.6] = 0
    predicted = pd.Series(predicted, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], predicted], axis=1)
    return combined
```

```python
sp500=sp500.dropna()
predictions=backtest(sp500,model,new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"],predictions['Predictions'])
```
